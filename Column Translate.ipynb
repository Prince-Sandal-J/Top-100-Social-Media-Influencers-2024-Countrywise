```python
import pandas as pd
import os

def translate_headers(file_path, translation_dict):
    """
    Translates non-English headers in a CSV file to English.

    Args:
        file_path: Path to the CSV file.
        translation_dict: A dictionary mapping non-English headers to their English equivalents.
    """

    df = pd.read_csv(file_path)

    # Rename columns using the translation dictionary
    df.rename(columns=translation_dict, inplace=True)

    # Write the modified DataFrame back to the CSV file
    df.to_csv(file_path, index=False)

def process_directory(directory_path, translation_dict):
    """
    Processes all CSV files in a directory and its subdirectories.

    Args:
        directory_path: Path to the directory containing CSV files.
        translation_dict: A dictionary mapping non-English headers to their English equivalents.
    """

    for root, _, files in os.walk(directory_path):
        for file in files:
            if file.endswith(".csv"):
                file_path = os.path.join(root, file)
                try:
                    translate_headers(file_path, translation_dict)
                    print("Successful")
                except Exception as e:
                    print(f"Error processing {file_path}: {e}")

# Example usage:
translation_dict = {
    # Add your translations here, considering potential variations
    "NOMBRE": "Name",
    "SEGUIDORES": "Followers",
    "TP": "ER",
    "PAÃS": "Country",
    "TEMA DE INFLUENCIA": "Topic of Influence",
    "ALCANCE POTENCIAL": "Potential Reach",
    "GUARDAR": "Save",
    "INVITAR A LA CAMPAÑA": "INVITE TO THE CAMPAIGN"
}

directory_path = "C:\\Users\\Admin\\Desktop\\Data Analytics\\Social Media Influencers\\Top 100 Influencers"
process_directory(directory_path, translation_dict)
```
